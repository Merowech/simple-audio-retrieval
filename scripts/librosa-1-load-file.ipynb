# Load Libraries

To install these libraries use: ```pip install numpy matplotlib librosa```

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (20,10)

import librosa
import librosa.display
````

# Load Basic Music File

I will use one of the FMA examples. I just randomly took the first one.

````python
# set the file path for a file you want to load
music_file = '../data/fma_small/000/000002.mp3'

# load our selected music file in librosa
y, sr = librosa.load(music_file)

# y is the time series extracted from the file
# sr is the sampling rate - you can also change this while loading but the default works in most cases
# the warning below just fails for PySoundFile as I didn't install it
````

````python
# we can take a closer look at the time series by jsut plotting it
plt.plot(y)
plt.show()
````
